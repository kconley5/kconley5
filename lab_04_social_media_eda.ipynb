## Prepare Data

➡️ Assignment Tasks
- Convert any numbers, such as quote_count, from strings into numbers
- Reduce the variables in your dataframe to only those that have useful data
- Optional: Locate outliers in any of the variables.   Filter them out if they are mistakes; keep them if they are exceptional cases.

```python
#Convert variables from strings to numeric
df['public_metrics.quote_count'] = pd.to_numeric(df['public_metrics.quote_count'])
df['public_metrics.reply_count'] = pd.to_numeric(df['public_metrics.reply_count'])
df['public_metrics.like_count'] = pd.to_numeric(df['public_metrics.like_count'])
df['public_metrics.retweet_count'] = pd.to_numeric(df['public_metrics.retweet_count'])
df['public_metrics.bookmark_count'] = pd.to_numeric(df['public_metrics.bookmark_count'])
df['public_metrics.impression_count'] = pd.to_numeric(df['public_metrics.impression_count'])

#print values to see type change
print(df['public_metrics.impression_count'])
print(df['public_metrics.quote_count'])
print(df['public_metrics.reply_count'])
print(df['public_metrics.like_count'])
print(df['public_metrics.retweet_count'])
print(df['public_metrics.bookmark_count'])
```

## Summarize Data

➡️ Assignment Tasks
- Create at least two charts to summarize and visualize the data
- Examples include: Number of posts per device; top 10 keywords with highest engagement

```python
#import library
import matplotlib.pyplot as plt

#Group by username and sum likes
username_likes = df.groupby('username')['public_metrics.like_count'].sum()

#Sort by likes in descending order and get the top 10
top_10_usernames = username_likes.sort_values(ascending=False).head(10)

#Create a bar chart
plt.bar(top_10_usernames.index, top_10_usernames.values)
plt.title("Top 10 Usernames with the Most Likes")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Sort the dataframe by impression count in descending order and get the top 20 posts
top_20_posts = df.sort_values('public_metrics.impression_count', ascending=False).head(20)

# Create the chart
plt.bar(top_20_posts['keyword'], top_20_posts['public_metrics.impression_count'])
plt.xlabel("Keyword")
plt.ylabel("Number of Impressions")
plt.title("Top 20 Posts: Impressions per Keyword")
plt.xticks(rotation=45, ha="right") # Rotate x-axis labels for better readability
plt.show()
```
